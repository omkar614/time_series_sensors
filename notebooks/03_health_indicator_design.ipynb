```python
def rolling_slope(x):
    if len(x)<2:
        return 0
    return np.polyfit(range(len(x)),x,1)[0]

for s in important_sensors:
    train[f"{s}_slope"]=train.groupby('id')[s].transform(
        lambda x:x.rolling(10,min_periods = 2).apply(rolling_slope,raw=False)
    )
```

```python
for s in important_sensors:
    train[f'{s}_norm'] = train.groupby('id')[s].transform(
        lambda x:(x-x.mean())/x.std()+1e-6
    )

norm_cols = [f'{s}_norm' for s in important_sensors]
train["health_score"] = 100 - (train[norm_cols].abs().mean(axis=1) * 10)
train["health_score"] = train["health_score"].clip(0, 100)
```

```python
feature_cols = (
    ["id", "cycle", "RUL", "health_score"] +
    [c for c in train.columns if "roll" in c or "slope" in c]
)

features = train[feature_cols]
features.to_csv(r"D:\projects\aviation\dataset\engine_health_features.csv", index=False)
```
